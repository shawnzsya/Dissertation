```python
def filter_outliers(dataframe, column_name, lower_bound=None, upper_bound=None):
    var = dataframe[column_name]
    if lower_bound is None and upper_bound is None:  # Use IQR method if bounds are not specified
        q1 = var.quantile(0.25)  # Calculate the first quartile
        q3 = var.quantile(0.75)  # Calculate the third quartile
        iqr = q3 - q1  # Calculate the interquartile range
        lower_bound = q1 - 3 * iqr  # Calculate the lower bound
        upper_bound = q3 + 3 * iqr  # Calculate the upper bound
    # Filter based on the specified bounds or the calculated IQR bounds, including the upper bound
    filtered = dataframe[(var >= lower_bound) & (var <= upper_bound)]
    dropped_observations = len(dataframe) - len(filtered)
    print('Dropped {} observations from {}'.format(dropped_observations, column_name))
    return filtered

# Set the range specifically for the review scores location column
df = filter_outliers(df, 'age_of_driver', lower_bound=18, upper_bound=80)


```

## Methodology

### K-means

```python
# 选择经纬度进行聚类
coords = geo_df[['longitude', 'latitude']].values
# 标准化特征
scaler = StandardScaler()
coords_scaled = scaler.fit_transform(coords)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import hdbscan
from sklearn.cluster import KMeans

# 初始HDBSCAN聚类
coords = np.array(geo_df[['longitude', 'latitude']])  # Assuming your coordinates are in these columns
clusterer = hdbscan.HDBSCAN(min_cluster_size=200, min_samples=70)
labels = clusterer.fit_predict(coords)

# 计算每个聚类的中心点
cluster_centers = []
for label in set(labels):
    if label != -1:  # 忽略噪声点
        cluster_points = coords[labels == label]
        cluster_center = cluster_points.mean(axis=0)
        cluster_centers.append(cluster_center)

# 将中心点转换为数组
cluster_centers = np.array(cluster_centers)

# 对中心点进行KMeans二次聚类
n_clusters = 5  # 目标聚类数量
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(cluster_centers)
second_labels = kmeans.labels_

# 创建一个新的标签数组，将原始标签映射到新的二次聚类标签
final_labels = np.full_like(labels, -1)
for original_label, new_label in zip(set(labels), second_labels):
    final_labels[labels == original_label] = new_label

# 将新的聚类结果添加到数据集中
geo_df['cluster'] = final_labels

# 删除噪声点
geo_df_no_noise = geo_df[geo_df['cluster'] != -1]

# 可视化二次聚类后删除噪声点的结果
fig, ax = plt.subplots(figsize=(10, 10))
london_map.plot(ax=ax, color='white', edgecolor='black')

unique_labels = set(geo_df_no_noise['cluster'])
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    class_member_mask = (geo_df_no_noise['cluster'] == k)
    xy = geo_df_no_noise[class_member_mask]
    ax.plot(xy['longitude'], xy['latitude'], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=5, label=f'Cluster {k}')

ax.set_title('Secondary Clustering Results without Noise')
ax.legend(loc='best', title="Clusters")

plt.savefig('./image/Secondary_no_noise_cluster.png', dpi=300, bbox_inches='tight')
plt.show()

# 统计每个聚类的数量
cluster_counts = geo_df_no_noise['cluster'].value_counts().sort_index()

# 打印每个聚类的数量
print("Number of points in each cluster after secondary clustering:")
for cluster, count in cluster_counts.items():
    print(f"Cluster {cluster}: {count} points")

# 导出去除噪声点后的数据
geo_df_no_noise.to_csv('road_collision_clusters_secondary_no_noise.csv', index=False)
```

```python
# calculating VIF
# This function is amended from: https://stackoverflow.com/a/51329496/4667568

from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, list_var_not_to_remove=None, thresh=5):
    '''
    Calculates VIF for each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param list_var_not_to_remove: the list of variables that should not be removed even though it has a high VIF. For example, dummy (or indicator) variables represent a categorical variable with three or more categories.
    :param thresh: the max VIF value before the feature is removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constant item to the data
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name="VIF",
              index=df_with_const.columns).to_frame()

        # drop the const as const should not be removed
        vif_df = vif_df.drop('const')
        
        # drop the variables that should not be removed
        if list_var_not_to_remove is not None:
            vif_df = vif_df.drop(list_var_not_to_remove)
            
        print('Current VIF values:')
        print(vif_df)
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns=index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

### XGBoost（全局---整个伦敦地区）

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

randomState_xgb = 100
randomState_dt = 42  # Ensure randomState_dt is defined

# Adjust class labels from 1, 2, 3 to 0, 1, 2
Y_resampled_adjusted_xgboost = Y_resampled - 1

# Split the data into training and testing sets
X_train_xgboost, X_test_xgboost, Y_train_xgboost, Y_test_xgboost = train_test_split(
    X_resampled, Y_resampled_adjusted_xgboost, test_size=0.2, random_state=randomState_dt
)

# Define hyperparameters
hyperparameters_xgboost = {'max_depth': [10, 20, 30, 40, 50], 'n_estimators': [100, 200, 300, 400, 500]}

# Initialize XGBoost classifier
xgb_xgboost = XGBClassifier(random_state=randomState_xgb)

# Use GridSearchCV for hyperparameter tuning
gscv_xgboost = GridSearchCV(xgb_xgboost, hyperparameters_xgboost, cv=5)
gscv_xgboost.fit(X_train_xgboost, Y_train_xgboost)

# Output the best parameters and value
print("The best parameter value is:")
print(gscv_xgboost.best_params_)
print("The best score is:")
print(gscv_xgboost.best_score_)

# Train the final model using the best parameters
best_xgb_model_xgboost = gscv_xgboost.best_estimator_

# Predict on the test set
Y_pred_xgboost = best_xgb_model_xgboost.predict(X_test_xgboost)
```

```python
df_cluster = df_clusters[df_clusters['cluster'] == 1]
longitude_latitude = df_cluster[['longitude', 'latitude']]
data_X = df_cluster.drop(['accident_severity', 'accident_index', 'accident_reference', 'date', 'time', 'geometry', 'neighbourhood', 'index_right', 'cluster'], axis=1)
data_Y = df_cluster['accident_severity']

randomState_dt = 100

# 使用 SMOTE 进行过采样
smote = SMOTE(random_state=42, k_neighbors=2)
X_resampled, Y_resampled = smote.fit_resample(data_X, data_Y)

# Define hyperparameters
hyperparameters_rf = {
    'n_estimators': [300, 400, 500, 600, 700],
    'max_depth': [10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy'],
}

# Convert X_resampled to DataFrame
X_resampled_rf = pd.DataFrame(X_resampled, columns=data_X.columns)
Y_resampled_rf = pd.Series(Y_resampled)

# Extract and merge longitude and latitude information
X_resampled_rf['longitude'] = longitude_latitude.loc[
    data_X.index
].values[X_resampled_rf.index % len(data_X)]
X_resampled_rf['latitude'] = longitude_latitude.loc[
    data_X.index
].values[X_resampled_rf.index % len(data_X)]

# Convert to GeoDataFrame
gdf_rf = gpd.GeoDataFrame(
    X_resampled_rf,
    geometry=gpd.points_from_xy(X_resampled_rf.longitude, X_resampled_rf.latitude),
)

# Divide the dataset into several spatial regions (e.g., grids)
gdf_rf['region'] = (
    (gdf_rf.geometry.x // 0.01).astype(int) * 1000 + (gdf_rf.geometry.y // 0.01).astype(int)
)


# Use spatial regions to split the training and test sets
train_gdf_rf, test_gdf_rf = train_test_split(
    gdf_rf, test_size=0.2, random_state=100, stratify=gdf_rf['region']
)

# Extract features and labels from GeoDataFrames
X_train_rf = train_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
Y_train_rf = Y_resampled_rf.loc[train_gdf_rf.index]
X_test_rf = test_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
Y_test_rf = Y_resampled_rf.loc[test_gdf_rf.index]


# Initialize Random Forest Classifier
rf = RandomForestClassifier(random_state=randomState_dt)

# Use GridSearchCV for hyperparameter tuning
clf_rf = GridSearchCV(rf, hyperparameters_rf, cv=5, n_jobs=-1)  # cv=5 for 5-fold cross-validation

# Train the model and find the best parameters
clf_rf.fit(X_train_rf, Y_train_rf)

print("The best parameter value is:")
print(clf_rf.best_params_)
print("The best score is:")
print(clf_rf.best_score_)

# Train the model using the best parameters (automatically done by GridSearchCV)
best_rf_model_rf = clf_rf.best_estimator_

# Predict on the test set
Y_pred_rf = best_rf_model_rf.predict(X_test_rf)
```

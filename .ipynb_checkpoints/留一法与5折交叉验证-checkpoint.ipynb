```python
from sklearn.model_selection import KFold,LeaveOneOut,LeavePOut,ShuffleSplit # 交叉验证所需的子集划分方法
loo = LeaveOneOut()
```

```python
# 定义超参数
hyperparameters = {
    'n_estimators': [300, 400, 500, 600, 700],
    'max_depth': [10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# 初始化随机森林分类器
rf = RandomForestClassifier(random_state=randomState_dt)

# 使用 GridSearchCV 进行超参数调节
clf = GridSearchCV(rf, hyperparameters, cv=5, n_jobs=-1)  # cv=5 for 5 fold cross verification

# 使用 GridSearchCV 进行超参数调节,loo交叉验证
# clf = GridSearchCV(rf, hyperparameters, cv=loo, n_jobs=-1)  # cv=5 for 5 fold cross verification

# 训练模型并找到最佳参数
clf.fit(X_train, Y_train)

print("The best parameter value is: ")
print(clf.best_params_)
print("The best score is: ")
print(clf.best_score_)

# 使用最佳参数训练模型（已经由 GridSearchCV 自动完成）
best_rf_model = clf.best_estimator_

# 在测试集上进行预测
Y_pred = best_rf_model.predict(X_test)
```

```python
# 生成混淆矩阵
conf_matrix = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix for RF:")
print(conf_matrix)

# 计算评估指标
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
recall = recall_score(Y_test, Y_pred, average='weighted')
f1 = f1_score(Y_test, Y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# 生成分类报告
class_report = classification_report(Y_test, Y_pred)
print("Classification Report  for RF:")
print(class_report)
```

```python
import shap
import numpy as np
import matplotlib.pyplot as plt

# 使用 SHAP 进行特征重要性分析
explainer = shap.TreeExplainer(best_rf_model, X_train, feature_perturbation='interventional')
shap_values = explainer(X_test, check_additivity=False)

# 获取每个类别的 SHAP 值
shap_values_class_1 = shap_values.values[:, :, 0]
shap_values_class_2 = shap_values.values[:, :, 1]
shap_values_class_3 = shap_values.values[:, :, 2]

# 计算每个特征的重要性平均值
shap_mean_class_1 = np.mean(np.abs(shap_values_class_1), axis=0)
shap_mean_class_2 = np.mean(np.abs(shap_values_class_2), axis=0)
shap_mean_class_3 = np.mean(np.abs(shap_values_class_3), axis=0)
```

```python
# 合并数据
# shap_mean_combined = np.vstack([shap_mean_class_1, shap_mean_class_2]).T
shap_mean_combined = np.vstack([shap_mean_class_1, shap_mean_class_2, shap_mean_class_3]).T
# 特征名
feature_names = X_test.columns

# 设置颜色和类别标签
colors = ['blue', 'red', 'green']
class_labels = ['Fatal', 'Serious', 'Slight']

# 绘制 Class-Specific Bar Plot
fig, ax = plt.subplots(figsize=(10, 8))
width = 0.2  # 柱的宽度
x = np.arange(len(feature_names))  # 特征的 x 轴位置

# 为每个类别绘制柱状图
for i in range(shap_mean_combined.shape[1]):
    ax.barh(x + width * i, shap_mean_combined[:, i], width, label=class_labels[i], color=colors[i])

# 添加特征名到 y 轴
ax.set_yticks(x + width)
ax.set_yticklabels(feature_names)

# 添加标签和标题
ax.set_xlabel("Mean(|SHAP value|) (average impact on model output magnitude)")
ax.set_title("SHAP Feature Importance by Class")
ax.legend()
plt.savefig(r'D:\Program File\Jupyter data\dataprocess\Dissertation\pictures\SHAP_分类_重要性类别总.png', dpi=300, bbox_inches='tight')
plt.show()

# 绘制 Overlay SHAP Feature Importance by Class
fig, ax = plt.subplots(figsize=(12, 8))

# 按特征重要性排序
sorted_idx = np.argsort(np.sum(shap_mean_combined, axis=1))[::-1]

# 累积宽度，叠加图形
for i in range(shap_mean_combined.shape[1]):
    ax.barh(feature_names[sorted_idx], shap_mean_combined[sorted_idx, i], label=class_labels[i], color=colors[i])

# 添加标签和标题
ax.set_xlabel("Mean(|SHAP value|) (average impact on model output magnitude)")
ax.set_title("Overlay layer SHAP Feature Importance by Class")
ax.legend()
plt.savefig(r'D:\Program File\Jupyter data\dataprocess\Dissertation\pictures\SHAP_分类_重要性图总.png', dpi=300, bbox_inches='tight')
plt.show()

# 计算所有类别 SHAP 值的平均值
shap_values_mean = np.mean(shap_values.values, axis=2)

# 创建一个新的 SHAP Explanation 对象，其中包含综合后的 SHAP 值
shap_values_combined = shap.Explanation(
    values=shap_values_mean,
    base_values=np.mean(shap_values.base_values, axis=1),  # 基线值取平均值
    data=X_test,
    feature_names=X_test.columns
)

# 绘制综合 beeswarm 图，设置 max_display 参数以显示所有特征
plt.figure(figsize=(10, 8))
shap.plots.beeswarm(shap_values_combined, max_display=shap_values_combined.values.shape[1], show=False)
plt.savefig(r'D:\Program File\Jupyter data\dataprocess\Dissertation\pictures\SHAP_分类_重要性热点图总.png', dpi=300, bbox_inches='tight')
plt.close()

# 绘制条形图以展示变量对模型的影响因素大小，设置 max_display 参数以显示所有特征
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_combined, X_test, plot_type="bar", max_display=shap_values_combined.values.shape[1], show=False)
plt.savefig(r'D:\Program File\Jupyter data\dataprocess\Dissertation\pictures\SHAP摘要图总.png', dpi=300, bbox_inches='tight')
plt.close()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
import pickle
import shap
import numpy as np
import matplotlib.pyplot as plt
import os

# 定义超参数
hyperparameters = {
    'n_estimators': [300, 400, 500, 600, 700],
    'max_depth': [10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10]
}

# 随机数种子
randomState_dt = 100

# 保存结果的列表
results_list = []

# 确保保存图像的目录存在
os.makedirs('./image1', exist_ok=True)

# 遍历每个聚类类别
for cluster in df_clusters['cluster'].unique():
    try:
        # 选择当前聚类类别的数据
        df_cluster = df_clusters[df_clusters['cluster'] == cluster]
        
        # 随机选择1000个样本（如果超过1000）
        if len(df_cluster) > 1000:
            df_cluster = df_cluster.sample(n=1000, random_state=randomState_dt)
        
        # 删除指定的列
        data_X = df_cluster.drop(['accident_severity', 'accident_index', 'accident_reference', 'date', 'time', 'geometry', 'neighbourhood', 'index_right', 'cluster','longitude','latitude'], axis=1)
        data_Y = df_cluster['accident_severity']
        
        # 使用 SMOTE 进行过采样，调整 k_neighbors 参数
        smote = SMOTE(random_state=42, k_neighbors=min(2, min(data_Y.value_counts()) - 1))
        X_resampled, Y_resampled = smote.fit_resample(data_X, data_Y)
        
        # 拆分训练集和测试集
        X_train, X_test, Y_train, Y_test = train_test_split(X_resampled, Y_resampled, test_size=0.2, random_state=randomState_dt)
        
        # 初始化随机森林分类器
        rf = RandomForestClassifier(random_state=randomState_dt)
        
        # 使用 GridSearchCV 进行超参数调节
        clf = GridSearchCV(rf, hyperparameters, cv=5, n_jobs=-1)  # cv=5 for 5-fold cross-validation
        
        # 训练模型
        clf.fit(X_train, Y_train)
        
        # 使用最佳参数训练模型并进行预测
        best_rf_model = clf.best_estimator_
        Y_pred = best_rf_model.predict(X_test)

        # 评估模型
        accuracy = accuracy_score(Y_test, Y_pred)
        report = classification_report(Y_test, Y_pred)
        precision = precision_score(Y_test, Y_pred, average='weighted')
        recall = recall_score(Y_test, Y_pred, average='weighted')
        f1 = f1_score(Y_test, Y_pred, average='weighted')

        # 打印训练和测试集上的 R² 值
        train_r2 = best_rf_model.score(X_train, Y_train)
        test_r2 = best_rf_model.score(X_test, Y_test)

        # 保存结果
        results = {
            'cluster': cluster,
            'best_params': clf.best_params_,
            'best_score': clf.best_score_,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'report': report,
            'train_r2': train_r2,
            'test_r2': test_r2,
            'best_rf_model': best_rf_model,
            'X_train': X_train,
            'X_test': X_test
        }
        
        # 添加到结果列表
        results_list.append(results)

        # 使用 SHAP 进行特征重要性分析
        explainer = shap.TreeExplainer(best_rf_model, X_train, feature_perturbation='interventional')
        shap_values = explainer(X_test, check_additivity=False)

        # 计算所有类别 SHAP 值的平均值
        shap_values_mean = np.mean(shap_values.values, axis=2)

        # 创建一个新的 SHAP Explanation 对象，其中包含综合后的 SHAP 值
        shap_values_combined = shap.Explanation(
            values=shap_values_mean,
            base_values=np.mean(shap_values.base_values, axis=1),  # 基线值取平均值
            data=X_test,
            feature_names=X_test.columns
        )

        # 绘制综合 beeswarm 图，设置 max_display 参数以显示所有特征
        plt.figure(figsize=(10, 8))
        shap.plots.beeswarm(shap_values_combined, max_display=shap_values_combined.values.shape[1], show=False)
        plt.savefig(f'./image/beeswarm_plot_cluster_{cluster}.png', dpi=300, bbox_inches='tight')
        plt.close()

        # 绘制条形图以展示变量对模型的影响因素大小，设置 max_display 参数以显示所有特征
        plt.figure(figsize=(10, 8))
        shap.summary_plot(shap_values_combined, X_test, plot_type="bar", max_display=shap_values_combined.values.shape[1], show=False)
        plt.savefig(f'./image/summary_plot_cluster_{cluster}.png', dpi=300, bbox_inches='tight')
        plt.close()
    
    except ValueError as e:
        print(f"Cluster {cluster} - Error: {e}")
        results_list.append({
            'cluster': cluster,
            'error': str(e)
        })

# 保存结果到文件
with open('results.pkl', 'wb') as f:
    pickle.dump(results_list, f)

# 输出所有聚类类别的结果
for result in results_list:
    print(f"Cluster {result['cluster']} Results:")
    for key, value in result.items():
        if key not in ['best_rf_model', 'X_train', 'X_test']:
            print(f"{key}: {value}")
    print("\n")
```

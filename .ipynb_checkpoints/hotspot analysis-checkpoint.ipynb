```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from imblearn.over_sampling import SMOTE
import pandas as pd
import geopandas as gpd
import shap  # SHAP for explainability
import matplotlib.pyplot as plt
import numpy as np

# Define a function to train and evaluate the model for a specific cluster
def evaluate_random_forest_with_shap(df_clusters, cluster_number):
    # Filter data for the specific cluster
    df_cluster = df_clusters[df_clusters['cluster'] == cluster_number]
    
    # Extract features and labels
    longitude_latitude = df_cluster[['longitude', 'latitude']]
    data_X = df_cluster.drop(
        ['accident_severity', 'accident_index', 'accident_reference', 'date', 'time', 'geometry', 'neighbourhood', 'index_right', 'cluster'], 
        axis=1
    )
    data_Y = df_cluster['accident_severity']

    randomState_dt = 100

    # 使用 SMOTE 进行过采样
    smote = SMOTE(random_state=42, k_neighbors=1)
    X_resampled, Y_resampled = smote.fit_resample(data_X, data_Y)

    # Define hyperparameters
    hyperparameters_rf = {
        'n_estimators': [300, 400, 500, 600, 700],
        'max_depth': [10, 20, 30, 40, 50],
        'min_samples_split': [2, 5, 10],
        'criterion': ['gini', 'entropy'],
    }

    # Convert X_resampled to DataFrame
    X_resampled_rf = pd.DataFrame(X_resampled, columns=data_X.columns)
    Y_resampled_rf = pd.Series(Y_resampled)

    # Extract and merge longitude and latitude information
    X_resampled_rf['longitude'] = longitude_latitude.loc[
        data_X.index
    ].values[X_resampled_rf.index % len(data_X)]
    X_resampled_rf['latitude'] = longitude_latitude.loc[
        data_X.index
    ].values[X_resampled_rf.index % len(data_X)]

    # Convert to GeoDataFrame
    gdf_rf = gpd.GeoDataFrame(
        X_resampled_rf,
        geometry=gpd.points_from_xy(X_resampled_rf.longitude, X_resampled_rf.latitude),
    )

    # Divide the dataset into several spatial regions (e.g., grids)
    gdf_rf['region'] = (
        (gdf_rf.geometry.x // 0.01).astype(int) * 1000 + (gdf_rf.geometry.y // 0.01).astype(int)
    )

    # Use spatial regions to split the training and test sets
    train_gdf_rf, test_gdf_rf = train_test_split(
        gdf_rf, test_size=0.2, random_state=100, stratify=gdf_rf['region']
    )

    # Extract features and labels from GeoDataFrames
    X_train_rf = train_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
    Y_train_rf = Y_resampled_rf.loc[train_gdf_rf.index]
    X_test_rf = test_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
    Y_test_rf = Y_resampled_rf.loc[test_gdf_rf.index]

    # Initialize Random Forest Classifier
    rf = RandomForestClassifier(random_state=randomState_dt)

    # Use GridSearchCV for hyperparameter tuning
    clf_rf = GridSearchCV(rf, hyperparameters_rf, cv=5, n_jobs=-1)  # cv=5 for 5-fold cross-validation

    # Train the model and find the best parameters
    clf_rf.fit(X_train_rf, Y_train_rf)

    print("The best parameter value is:")
    print(clf_rf.best_params_)
    print("The best score is:")
    print(clf_rf.best_score_)

    # Train the model using the best parameters (automatically done by GridSearchCV)
    best_rf_model_rf = clf_rf.best_estimator_

    # Predict on the test set
    Y_pred_rf = best_rf_model_rf.predict(X_test_rf)

    # Calculate evaluation metrics
    accuracy_rf = accuracy_score(Y_test_rf, Y_pred_rf)
    precision_rf = precision_score(Y_test_rf, Y_pred_rf, average='weighted')
    recall_rf = recall_score(Y_test_rf, Y_pred_rf, average='weighted')
    f1_rf = f1_score(Y_test_rf, Y_pred_rf, average='weighted')

    # Display evaluation metrics
    print("Evaluation Metrics for Random Forest:")
    print(f"Accuracy: {accuracy_rf:.4f}")
    print(f"Precision: {precision_rf:.4f}")
    print(f"Recall: {recall_rf:.4f}")
    print(f"F1 Score: {f1_rf:.4f}")

    # Display classification report
    print("\nClassification Report:")
    print(classification_report(Y_test_rf, Y_pred_rf))
    
    # SHAP Analysis
    # 使用 SHAP 进行特征重要性分析
    explainer = shap.TreeExplainer(best_rf_model_rf, X_train_rf, feature_perturbation='interventional')
    shap_values = explainer(X_test_rf, check_additivity=False)
    # 重新计算综合后的 SHAP 值
    shap_values_mean = np.mean(np.abs(shap_values.values), axis=2)

      # 创建一个新的 SHAP Explanation 对象，其中包含综合后的 SHAP 值
    shap_values_combined = shap.Explanation(
        values=shap_values_mean,
        base_values=np.mean(shap_values.base_values, axis=1),  # 基线值取平均值
        data=X_test_rf,  # Use X_test_rf instead of X_test
        feature_names=X_test_rf.columns
    )

    # Plot SHAP bar plot
    # 绘制条形图以展示变量对模型的影响因素大小，设置 max_display 参数以显示所有特征
    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_values_combined, X_test_rf, plot_type="bar", max_display=shap_values_combined.values.shape[1], show=False)
    plt.title(f"Summary Plot for Cluster {cluster_number}")
    plt.savefig(f'./image/summary_plot_cluster_{cluster_number}.png', dpi=300, bbox_inches='tight')
    plt.close()

# Example usage for cluster 1
evaluate_random_forest_with_shap(df_clusters, cluster_number=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from imblearn.over_sampling import SMOTE, RandomOverSampler
import shap

# Filter data for the specific cluster
df_cluster = df_clusters[df_clusters['cluster'] == 4]

# Extract features and labels
longitude_latitude = df_cluster[['longitude', 'latitude']]
data_X = df_cluster.drop(
    [
        'accident_severity', 
        'accident_index', 
        'accident_reference', 
        'date', 
        'time', 
        'geometry', 
        'neighbourhood', 
        'index_right', 
        'cluster'
    ], 
    axis=1
)
data_Y = df_cluster['accident_severity']

randomState_dt = 100

# Check the class distribution
class_distribution = data_Y.value_counts()
print("Class distribution before SMOTE:")
print(class_distribution)

# Use SMOTE for oversampling if possible
try:
    smote = SMOTE(random_state=42, k_neighbors=1)
    X_resampled, Y_resampled = smote.fit_resample(data_X, data_Y)
except ValueError as e:
    print(f"SMOTE failed due to: {e}")
    # Use RandomOverSampler as an alternative
    print("Using RandomOverSampler instead.")
    ros = RandomOverSampler(random_state=42)
    X_resampled, Y_resampled = ros.fit_resample(data_X, data_Y)

# Define hyperparameters for Random Forest
hyperparameters_rf = {
    'n_estimators': [300, 400, 500, 600, 700],
    'max_depth': [10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy'],
}

# Convert X_resampled to DataFrame
X_resampled_rf = pd.DataFrame(X_resampled, columns=data_X.columns)
Y_resampled_rf = pd.Series(Y_resampled)

# Extract and merge longitude and latitude information
X_resampled_rf['longitude'] = longitude_latitude.loc[
    data_X.index
].values[X_resampled_rf.index % len(data_X)]
X_resampled_rf['latitude'] = longitude_latitude.loc[
    data_X.index
].values[X_resampled_rf.index % len(data_X)]

# Convert to GeoDataFrame
gdf_rf = gpd.GeoDataFrame(
    X_resampled_rf,
    geometry=gpd.points_from_xy(X_resampled_rf.longitude, X_resampled_rf.latitude),
)

# Divide the dataset into several spatial regions (e.g., grids)
gdf_rf['region'] = (
    (gdf_rf.geometry.x // 0.01).astype(int) * 1000 + (gdf_rf.geometry.y // 0.01).astype(int)
)

# Use spatial regions to split the training and test sets
train_gdf_rf, test_gdf_rf = train_test_split(
    gdf_rf, test_size=0.2, random_state=randomState_dt, stratify=gdf_rf['region']
)

# Extract features and labels from GeoDataFrames
X_train_rf = train_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
Y_train_rf = Y_resampled_rf.loc[train_gdf_rf.index]
X_test_rf = test_gdf_rf.drop(['region', 'geometry', 'longitude', 'latitude'], axis=1)
Y_test_rf = Y_resampled_rf.loc[test_gdf_rf.index]

# Initialize Random Forest Classifier
rf = RandomForestClassifier(random_state=randomState_dt)

# Use GridSearchCV for hyperparameter tuning
clf_rf = GridSearchCV(rf, hyperparameters_rf, cv=5, n_jobs=-1)  # cv=5 for 5-fold cross-validation

# Train the model and find the best parameters
clf_rf.fit(X_train_rf, Y_train_rf)

print("The best parameter value is:")
print(clf_rf.best_params_)
print("The best score is:")
print(clf_rf.best_score_)

# Train the model using the best parameters (automatically done by GridSearchCV)
best_rf_model_rf = clf_rf.best_estimator_

# Predict on the test set
Y_pred_rf = best_rf_model_rf.predict(X_test_rf)

# Calculate evaluation metrics
accuracy_rf = accuracy_score(Y_test_rf, Y_pred_rf)
precision_rf = precision_score(Y_test_rf, Y_pred_rf, average='weighted')
recall_rf = recall_score(Y_test_rf, Y_pred_rf, average='weighted')
f1_rf = f1_score(Y_test_rf, Y_pred_rf, average='weighted')

# Display evaluation metrics
print("Evaluation Metrics for Random Forest:")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1 Score: {f1_rf:.4f}")

# Display classification report
print("\nClassification Report:")
print(classification_report(Y_test_rf, Y_pred_rf))

# SHAP Analysis
# Use SHAP for feature importance analysis
explainer = shap.TreeExplainer(best_rf_model_rf, X_train_rf, feature_perturbation='interventional')
shap_values = explainer(X_test_rf, check_additivity=False)

# Recompute comprehensive SHAP values
shap_values_mean = np.mean(np.abs(shap_values.values), axis=2)

# Create a new SHAP Explanation object with comprehensive SHAP values
shap_values_combined = shap.Explanation(
    values=shap_values_mean,
    base_values=np.mean(shap_values.base_values, axis=1),  # Take average of baseline values
    data=X_test_rf,  # Use X_test_rf instead of X_test
    feature_names=X_test_rf.columns
)

# Plot SHAP bar plot
# Create a bar chart to show the influence of variables on the model, setting max_display to show all features
plt.figure(figsize=(10, 8))
shap.summary_plot(
    shap_values_combined, 
    X_test_rf, 
    plot_type="bar", 
    max_display=shap_values_combined.values.shape[1], 
    show=False
)
plt.title(f"Summary Plot for Cluster 4")
plt.savefig(f'./image/summary_plot_cluster_4.png', dpi=300, bbox_inches='tight')
```
